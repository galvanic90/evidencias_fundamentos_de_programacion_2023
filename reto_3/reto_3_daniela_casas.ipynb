```python
'''
Calculadora de para gestión y control de cultivos
Presentado por: Daniela Casas Vasquez
              CC. 1035440224

Psd: Los datos que presento para las etapas de los cultivos son ficticios
los uso para darle más sentido y cuerpo al problema
'''

# Definición de los tipos de cultivos, sus tiempos y etapas
cultivos = [
    {
        'nombre': 'Cultivo de arroz',
        'mantenimiento': 'Lunes, Jueves 4:00 pm',
        'regado': 'Lunes, Jueves 4:00 pm',
        'abono': 'Lunes, Jueves 4:00 pm',
        'etapas': ['Germinación: 0 - 15 días', 'Crecimiento: 15 - 60 días',
                   'Reproducción: 60 - 180 días', 'Cosecha: 180 - 210 días']
    },
    {
        'nombre': 'Cultivo de plátano',
        'mantenimiento': 'Martes, Viernes 1:00 pm',
        'regado': 'Martes, Viernes 1:00 pm',
        'abono': 'Martes, Viernes 1:00 pm',
        'etapas': ['Germinación: 0 - 7 días', 'Crecimiento: 7 - 45 días',
                   'Reproducción: 45 - 120 días', 'Cosecha: 120 - 150 días']
    },
    {
        'nombre': 'Cultivo de uva',
        'mantenimiento': 'Lunes, Sábado 7:00 am',
        'regado': 'Lunes, Sábado 7:00 am',
        'abono': 'Lunes, Sábado 7:00 am',
        'etapas': ['Germinación: 0 - 30 días', 'Crecimiento: 30 - 90 días',
                   'Reproducción: 90 - 210 días', 'Cosecha: 210 - 255 días']
    }
]

# Función para mostrar el menú principal

def mostrar_menu():
    print("1. Horario de gestión de cultivo")
    print("2. Etapas del cultivo")
    print("3. Información contable")
    print("4. Salir")

# Función para mostrar las opciones de cultivos (submenú)

def mostrar_opciones_cultivos():
    for idx, cultivo in enumerate(cultivos, start=1):
        print(f"{idx}. {cultivo['nombre']}")

# Función para mostrar los horarios de gestión de un cultivo (submenú)

def mostrar_horarios_gestion(cultivo):
    print(f"Horario de mantenimiento: {cultivo['mantenimiento']}")
    print(f"Horario de regado: {cultivo['regado']}")
    print(f"Horario de abono: {cultivo['abono']}")

# Función para mostrar las etapas de un cultivo (submenú)

def mostrar_etapas(cultivo):
    print("Etapas del cultivo:")
    for etapa in cultivo['etapas']:
        print(etapa)

# Función para calcular información contable (submenú)

def calcular_informacion_contable():
    # Ingresar gastos variables

    medicamentos = float(input("Ingrese gastos en medicamentos: "))
    imprevistos = float(input("Ingrese gastos en imprevistos: "))

    # Ingresar costos fijos

    mano_obra = float(input("Ingrese el costo de la mano de obra: "))
    abono = float(input("Ingrese el costo del abono: "))
    agua = float(input("Ingrese el costo de agua: "))
    mantenimiento = float(input("Ingrese el costo de mantenimiento: "))

    # Ingresar producción

    valor_arroba = float(input("Ingrese el valor de una arroba (12.5 kg): "))
    kilos_recolectados = float(input("Ingrese la cantidad de kilos recolectados: "))

    # Operaciones para calcular los costos y ganancias

    costos_totales_mes = medicamentos + imprevistos + mano_obra + abono + agua + mantenimiento
    costos_totales_mano_obra = mano_obra
    meses_sin_gastos = []  # Rellenar según los meses sin gastos
    meses_gasto_menor_100000 = []  # Rellenar según los meses con gasto menor a 100000
    meses_gasto_mayor_costo = []  # Rellenar según los meses con gasto mayor al costo total
    costo_promedio_fijos = (mano_obra + abono + agua + mantenimiento) / 4
    costo_promedio_variables = (medicamentos + imprevistos) / 2
    ganancia = valor_arroba * (kilos_recolectados / 12.5) - costos_totales_mes
    ganancia_incremento_precio = (valor_arroba * 1.37) * (kilos_recolectados / 12.5) - costos_totales_mes
    ganancia_disminucion_costos = (valor_arroba * (kilos_recolectados / 12.5)) * 0.63 - (costos_totales_mes * 0.95)

    # Reporte de información financiera

    print("Informe económico:")
    print(f"Costos totales del cultivo por mes: {costos_totales_mes:.2f}")
    print(f"Costos totales de mano de obra: {costos_totales_mano_obra:.2f}")
    print(f"Meses sin gastos: {', '.join(meses_sin_gastos)}")
    print(f"Meses con gasto menor a 100000: {', '.join(meses_gasto_menor_100000)}")
    print(f"Meses con costo total mayor al gasto total: {', '.join(meses_gasto_mayor_costo)}")
    print(f"Valor promedio de costos fijos: {costo_promedio_fijos:.2f}")
    print(f"Valor promedio de costos variables: {costo_promedio_variables:.2f}")
    print(f"¿Hubo ganancia? {'SI' if ganancia > 0 else 'NO'}")
    print(f"Ganancia: {ganancia:.2f}")
    print(f"Ganancia con incremento de precio: {ganancia_incremento_precio:.2f}")
    print(f"Ganancia con disminución de costos y producción: {ganancia_disminucion_costos:.2f}")

# Creación del menú principal
while True:
    mostrar_menu()
    opcion = input("Ingrese la opción deseada: ")

    if opcion == '1':
        mostrar_opciones_cultivos()
        opcion_cultivo = int(input("Seleccione el cultivo: ")) - 1
        mostrar_horarios_gestion(cultivos[opcion_cultivo])
    elif opcion == '2':
        mostrar_opciones_cultivos()
        opcion_cultivo = int(input("Seleccione el cultivo: ")) - 1
        mostrar_etapas(cultivos[opcion_cultivo])
    elif opcion == '3':
        calcular_informacion_contable()
    elif opcion == '4':
        print("¡Hasta luego!")
        break
    else:
        print("Opción inválida. Por favor, seleccione una opción válida.")
```

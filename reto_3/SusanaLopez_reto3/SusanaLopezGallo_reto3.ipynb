```python
#variables para los cultivos
cultivos = {
    "LECHUGAS": {
        "Mantenimiento": "Lunes, 4:00 pm",
        "Regado": "Martes, Miercoles, Viernes, 7:00 am",
        "Fertilizacion": "Miércoles, 8:00 am",
        "Etapas": ["vegetativo"]  },
    "TOMATES": {
        "Mantenimiento": "Lunes, viernes, 6:00 am",
        "Regado": "Martes, Jueves, 7:00 pm",
        "Fertilizacion": "Viernes, 2:00 pm",
        "Etapas": ["vegetativa", "reproductiva"]  },  }


#menú principal
def imprimir_menu():
    print("BIENVENIDO A LA HUERTA ESCOLAR")
    print("\n--- Menú Principal ---")
    print("1. Horario de Gestión de Cultivo")
    print("2. Etapas del Cultivo")


while True:
    imprimir_menu()
    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        print("\n--- Horario de Gestión de Cultivo ---")
        for i, cultivo in enumerate(cultivos.keys()):
            print(f"{i + 1}. {cultivo}")

        seleccion = int(input("Seleccione un cultivo: "))
        seleccion -= 1
        if 0 <= seleccion < len(cultivos):
            cultivo_seleccionado = list(cultivos.keys())[seleccion]
            info_cultivo = cultivos[cultivo_seleccionado]
            print(f"\nHorarios de {cultivo_seleccionado}:")
            print(f"Mantenimiento: {info_cultivo['Mantenimiento']}")
            print(f"Regado: {info_cultivo['Regado']}")
            print(f"Fertilizacion: {info_cultivo['Fertilizacion']}")
        else:
            print("Selección de cultivo no válida.")

    elif opcion == "2":
     print("\n--- Etapas del Cultivo ---")
     cultivo_seleccionado = input("Ingrese el nombre del cultivo: ")

    if cultivo_seleccionado in cultivos:
        etapas_cultivo = cultivos[cultivo_seleccionado]["Etapas"]
        print("Etapas del cultivo:")
        for etapa in etapas_cultivo:
            print(etapa)
    else:
        print("Cultivo no encontrado.")



```

```python
# Variables para información contable
costos = {
    "Mano de obra": 50000,
    "Abono": 45000,
    "Agua": 15000,
    "Mantenimiento": 50000,
}

valor_arroba = 58500
kilos_recolectados = 58000

# Función para calcular los costos totales
def calcular_costos_totales():
    return costos["Mano de obra"] + costos["Abono"] + costos["Agua"] + costos["Mantenimiento"]

# Función para imprimir el menú principal
def imprimir_menu():
    print("Bienvenido a la parte contable cultivo de arroz")
    print("3. Información Contable")
    print("4. Salir")

# Función para imprimir el submenú de costos
def imprimir_submenu_costos():
    print("\n--- Sub Menú de Costos ---")
    print("1. Mano de obra")
    print("2. Abono")
    print("3. Agua")
    print("4. Mantenimiento")

while True:
    imprimir_menu()
    opcion = input("Seleccione una opción: ")

    if opcion == "3":
        print("\n--- Información Contable ---")
        imprimir_submenu_costos()

        subopcion = input("Seleccione una opción: ")

        if subopcion == "1":
            costos["Mano de obra"] = float(input("Ingrese el costo de mano de obra: "))
        elif subopcion == "2":
            costos["Abono"] = float(input("Ingrese el costo de abono: "))
        elif subopcion == "3":
            costos["Agua"] = float(input("Ingrese el costo de agua: "))
        elif subopcion == "4":
            costos["Mantenimiento"] = float(input("Ingrese el costo de mantenimiento: "))
        else:
            print("Selección no válida.")

        valor_arroba = float(input("Ingrese el valor de la arroba de arroz: "))
        kilos_recolectados = float(input("Ingrese la cantidad de kilos recolectados: "))

        # Cálculos
        costos_totales = calcular_costos_totales()
        promedio_costos_fijos = costos_totales / len(cultivos)
        ingresos = valor_arroba * (kilos_recolectados / 12.5)
        ganancia = ingresos - costos_totales

        print("\n--- Informe Económico ---")
        print(f"Costos Totales del Cultivo: ${costos_totales}")
        print(f"Valor Promedio de Costos Fijos: ${promedio_costos_fijos}")
        print(f"¿Hubo Ganancia? {'SI' if ganancia > 0 else 'NO'}")
        print(f"Ganancia: ${ganancia}")

        precio_incrementado = valor_arroba * 1.37
        ganancia_incrementada = precio_incrementado * (kilos_recolectados / 12.5) - costos_totales

        costos_disminuidos = costos_totales * 0.95
        produccion_disminuida = kilos_recolectados * 0.63
        ganancia_disminuida = valor_arroba * (produccion_disminuida / 12.5) - costos_disminuidos

        print(f"Si el precio del kilo de arroz se incrementa en un 37%, la ganancia sería: ${ganancia_incrementada}")
        print(f"Si los costos y gastos se disminuyen en un 5% y la cantidad de arrobas producidas en un 63%, la ganancia sería: ${ganancia_disminuida}")

    elif opcion == "4":
        print("Saliendo de la aplicación.")
        break
    else:
        print("Selección no válida.")
```

```python
#	Crear un objeto de tipo ndarray, con números enteros entre 64 hasta 1024.
import numpy as np
arreglo = np.arange(64 , 1025)
arreglo
```

```python
# Redimensionar a una matriz bidimensional que tenga las 1024 observaciones, pero que su dimensión tenga tanto el mismo número de filas como de columnas, con la finalidad de obtener una cuadrada.
arreglo = np.arange(64 , 1025).reshape((31,31))
arreglo
```

```python
# crear columna de datos aleatorios de 30 observaciones

DatosAleatorios = np.random.randint(1,100,10)
Seleccion_datos=[]
for i in range(0,10,1):
  elegido = np.random.choice(DatosAleatorios)
  Seleccion_datos.append(elegido)

Seleccion = np.array(Seleccion_datos).reshape(10,1)
Seleccion

```

```python
# creando cedulas con numeros aleatorios
Lista = ["Andres" , "Maria" , "Manuel" , "Daniel" , "Sarah" , "Cristian" , "Violetta" , "Lucia" , "Jackson" , "Jose"]
arreglo = np.array(Lista)
arreglo

cedulaAleatoria = np.random.randint(1000,9999,100)
cedulas=[]
for i in range(0,10,1):
  elegido = np.random.choice(cedulaAleatoria)
  cedulas.append(elegido)

cedulas
```

```python
# concatenar vector con las observaciones de nmbres y cedulas
np.ma.concatenate(([arreglo], [cedulas]), axis=0)
```

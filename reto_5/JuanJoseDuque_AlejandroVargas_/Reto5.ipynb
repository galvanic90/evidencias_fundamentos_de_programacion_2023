```python
##Reto 5
##Juan José Duque
##Alejandro Vargas

# Crear una lista para almacenar los estudiantes
lista_estudiantes = []
# Variable global gana, pierde
gana= 3.0
# Función para agregar un estudiante
def agregar_estudiante():
    print("Ingresa los datos del estudiante:")
    estudiante = {}
    estudiante["Identificación"] = input("Identificación: ")
    estudiante["Nombre"] = input("Nombre: ")
    estudiante["Correo"] = input("Correo: ")
    estudiante["Teléfono"] = input("Teléfono: ")
    estudiante["Fecha de nacimiento"] = input("Fecha de nacimiento: ")
    estudiante["Nota 1"] = float(input("Nota 1: "))
    estudiante["Nota 2"] = float(input("Nota 2: "))
    estudiante["Nota 3"] = float(input("Nota 3: "))
    estudiante["Nota 4"] = float(input("Nota 4: "))
    lista_estudiantes.append(estudiante)
    print("Estudiante agregado correctamente.")

# Función para calcular el promedio de notas de un estudiante
def calcular_promedio(notas):
  global gana
  return sum(notas) / len(notas)
  promedio= calcular_promedio(estudiante)
  print("Promedio de Nota ", promedio)
  if promedio >= gana:
      print("Gana")
  else:
      print("Pierde")

# Función para mostrar los datos de un estudiante
def mostrar_estudiante(estudiante):
    print("\nDatos del estudiante:")
    print("Identificación:", estudiante["Identificación"])
    print("Nombre:", estudiante["Nombre"])
    print("Correo:", estudiante["Correo"])
    print("Teléfono:", estudiante["Teléfono"])
    print("Fecha de nacimiento:", estudiante["Fecha de nacimiento"])
    print("Notas:")
    for i in range(1, 5):
        print(f"Nota {i}: {estudiante['Nota ' + str(i)]}")
 # Función para modificar las notas de un estudiante
def modificar_notas(estudiante):
    print("Notas actuales del estudiante:")
    for i in range(1, 5):
        print(f"Nota {i}: {estudiante['Nota ' + str(i)]}")

    for i in range(1, 5):
        nueva_nota = float(input(f"Nueva nota {i}: "))
        estudiante["Nota " + str(i)] = nueva_nota
        print("Nota Modificadas ")
# Función para calcular promedio
def calcular_promedio_total():
  if len(lista_estudiantes) == 0:
    return 0.0
  total_notas = sum(sum(estudiante[f'Nota {i}'] for i in range(1, 5)) for estudiante in lista_estudiantes)
  cantidad_notas = len(lista_estudiantes) * 4
  promedio_total = total_notas / cantidad_notas
  return promedio_total

# Menú principal
while True:
    print("\nMenú Principal:")
    print("1. Agregar Estudiante")
    print("2. Calcular Promedio de Notas")
    print("3. Mostrar Datos de un Estudiante")
    print("4. Modificar Notas de un Estudiante")
    print("5. Promedios")
    print("6. Salir")
    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_estudiante()
    elif opcion == "2":
        if len(lista_estudiantes) == 0:
            print("No hay estudiantes registrados.")
        else:
            identificacion = input("Ingrese la identificación del estudiante: ")
            for estudiante in lista_estudiantes:
                if estudiante["Identificación"] == identificacion:
                    notas = [estudiante["Nota 1"], estudiante["Nota 2"], estudiante["Nota 3"], estudiante["Nota 4"]]
                    promedio = calcular_promedio(notas)
                    print("Promedio de Nota ", promedio)
                    if promedio >= gana:
                      print("El Estudiante Gana la Materia")
                    else:
                      print("El Estudiante Pierde la Materia")
                    print(f"El promedio de notas del estudiante {estudiante['Nombre']} es: {promedio}")
                    break
    elif opcion == "3":
        if len(lista_estudiantes) == 0:
            print("No hay estudiantes registrados")
        else:
            identificacion = input("Ingrese la identificación del estudiante: ")
            for estudiante in lista_estudiantes:
                if estudiante["Identificación"] == identificacion:
                    mostrar_estudiante(estudiante)
                    break

    elif opcion == "4":
      if len(lista_estudiantes) == 0:
        print("No hay estudiantes registrados")
      else:
          identificacion = input("Ingrese la identificación del Estudiante: ")
          for estudiante in lista_estudiantes:
                if estudiante["Identificación"] == identificacion:
                    modificar_notas(estudiante)
                    break
    elif opcion == "5":
        promedio_total = calcular_promedio_total()
        print("Promedio de Notas Total:", promedio_total)
    elif opcion == "6":
        print("Saliendo del programa")
        break
```

```python
## defino clase estudiante

class estudiante:
   def __init__(self, ID, Nombre, Correo, Telefono, Fecha_Nacimiento, Nota1, Nota2, Nota3, Nota4):
      self.ID = ID
      self.Nombre = Nombre
      self.Correo = Correo
      self.Telefono = Telefono
      self.Fecha_Nacimiento = Fecha_Nacimiento
      self.Nota1 = Nota1
      self.Nota2 = Nota2
      self.Nota3 = Nota3
      self.Nota4 = Nota4
```

```python
## bloque de menú

def menu_options():
 a = int(input('''Elija la opcion deseada:\n
      [1] - Agregar
      [2] - Buscar
      [3] - Modificar
      [4] - Cancelar Materia
      [5] - Resultados por Estudiante
      [6] - Informe del Grupo
      [7] - Salir \n '''))
 return a
```

```python
##bloque de funciones que se llaman en el código principal
#funcion agregar
registro = dict()

def agregar():
    a = input("Identificacion : ")
    b = input("Nombre : ")
    c = input("Correo : ")
    d = input("Telefono : ")
    e = input("Fecha de nacimiento : ")
    f = input("Nota 1 : ")
    g = input("Nota 2 : ")
    h = input("Nota 3 : ")
    i = input("Nota 4 : ")
    registro[a] = estudiante(a,b,c,d,e,f,g,h,i)

#funcion buscar

def buscar():
    search = input("Ingrese la identificación del estudiante que desea consultar: ")

    if search in registro.keys():
      print(f"\n Estudiante : {registro[search].Nombre} \n Email: {registro[search].Correo} \n Tel: {registro[search].Telefono} \n Nota 1: {registro[search].Nota1} \n Nota 2: {registro[search].Nota2} \n Nota 3: {registro[search].Nota3} \n Nota 4: {registro[search].Nota4}")
    else:
      print(f"No existe un estudiante con ID : {search}")

#funcion modificar

def modificar():
    search = input("Ingrese la identificación del estudiante que desea modificar: ")

    if search in registro.keys():
      registro[search].Nota1 = input("ingrese nueva Nota 1: ")
      registro[search].Nota2 = input("ingrese nueva Nota 2: ")
      registro[search].Nota3 = input("ingrese nueva Nota 3: ")
      registro[search].Nota4 = input("ingrese nueva Nota 4: ")
    else:
      print(f"No existe un estudiante con ID = {search}")

#funcion eliminar

def eliminar():
  cancel = True

  while cancel == True:
    search = input("Ingrese la identificación del estudiante que desea dar de baja: ")

    if search in registro.keys():
      confirm = input(f"Desea borrar el estudiante con ID: {search} \n   (si / no): ")
      if confirm == "si":
        x = registro.pop(search)
        print(f"Estudiante con ID: {search} fue dado de baja")
        cancel = False
      else: pass
    else:
      print(f"No existe un estudiante con ID = {search}")

#funciones para calcular el resultado de un estudiante
#calcular promedio estudiante
def promedio_est(search):
  prom = (float(registro[search].Nota1) + float(registro[search].Nota2) + float(registro[search].Nota3) + float(registro[search].Nota4)) / 4
  return prom

#calcular promedio del grupo
def promedio_grupo():
  prom = 0
  for i in registro.keys():
   prom += promedio_est(i)
  return prom/len(registro)

#resultado de un estudiante
def resultado_est():

  search = input("Ingrese la identificación del estudiante que desea consultar: ")

  if search in registro.keys():
    prom_est = promedio_est(search)
    prom_gru = promedio_grupo()

    print(f"El promedio del estudiante es {prom_est}")
    print(f"Promedio Grupo : {prom_gru}")

    if prom_est > prom_gru: print("Estudiante sobre la media del grupo")
    else: print("Estudiante por debajo o igual a la media del grupo")

    if prom_est >= 3: print("Estudiante aprueba el curso")
    else: print("Estudiante reprueba el curso")

  else:
    print(f"No existe un estudiante con ID = {search}")

#----------#
#funcion informe del grupo

import numpy as np
from scipy import stats

def informe_grupo():
  sobre = 0
  debajo = 0
  igual = 0
  gana = 0
  pierde = 0
  lista_prom = []

  for i in registro.keys():
    if promedio_est(i) > promedio_grupo(): sobre += 1
    if promedio_est(i) == promedio_grupo(): igual += 1
    if promedio_est(i) < promedio_grupo(): debajo += 1
    if promedio_est(i) >= 3: gana += 1
    if promedio_est(i) < 3: pierde += 1

    lista_prom.append(promedio_est(i))
    print(f"-Estudiante ID: {registro[i].ID} -Nota final: {promedio_est(i)}")

  print(f"Estudiantes sobre la media : {sobre}")
  print(f"Estudiantes igual a la media : {igual}")
  print(f"Estudiantes debajo de la media : {debajo}")
  print(f"Estudiantes que aprueban : {gana} ({float(gana/len(registro))*100}%)")
  print(f"Estudiantes que reprueban : {pierde} ({float(pierde/len(registro))*100}%)")
  print(f"El promedio del grupo fue de: {promedio_grupo()}")

  print(f"La moda del grupo fue : {stats.mode(lista_prom)}")
  print(f"La mediana del grupo fue : {np.median(np.sort(lista_prom))}")
  print(f"La desviacion estándar del grupo fue : {np.std(lista_prom)}")
```

```python
#codigo principal que llama las funcions según la opción del menú capturada por consola

show_menu = True
# se hace en ciclo infinito para que retorne al menu cuando después de ejecutar una funcion, solo se cierra usando la opción 7
while show_menu == True:

  option = menu_options()

  if option == 1: agregar()
  elif option == 2: buscar()
  elif option == 3: modificar()
  elif option == 4: eliminar()
  elif option == 5: resultado_est()
  elif option == 6: informe_grupo()
  elif option == 7:
    print('''
    ¡Hasta pronto!
    ''')
    break

  else: print("opcion invalida")
```

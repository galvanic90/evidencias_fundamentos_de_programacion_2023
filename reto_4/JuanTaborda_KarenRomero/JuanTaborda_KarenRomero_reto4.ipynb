```python
## Bloque de funciones que contiene las opciones de cada menu

#menu principal
def menu_options():
  a = int(input('''Elije un tipo de operacion:
  \n [1] - basica
  \n [2] - extendida
  \n [3] - trigonometrica
  \n [4] - estadistica
  \n oprima otro número para salir \n''' ))
  return a

#menu de operaciones basicas
def basic_options():
  a = int(input('''Elija la operacion:
  \n [1] - Suma
  \n [2] - Resta
  \n [3] - Multiplicacion
  \n [4] - Division
  \n [5] - Regresar al menu principal \n'''))
  return a

#menu de operaciones extendidas
def extended_options():
 a = int(input('''Elija la operacion:
      \n [1] - division entera
      \n [2] - residuo
      \n [3] - exponenciacion
      \n [4] - raiz
      \n [5] - logaritmo
      \n [6] - logaritmo base10
      \n [7] - inversa
      \n [8] - factorial
      \n [9] - valor absoluto
      \n [10] - Regresar al menu principal \n'''))
 return a

#menu de operaciones trigonometricas
def trigo_options():
  a = int(input('''Elije un tipo de operacion trigonométrica:
  \n [1] - seno
  \n [2] - tan
  \n [3] - coseno
  \n [4] - Regresar al menu principal \n''' ))
  return a

#menu de operaciones estadisticas
def estadistica_options():
  a = int(input('''Elije un tipo de operacion estadística:
  \n [1] - promedio
  \n [2] - mediana
  \n [3] - moda
  \n [4] - Regresar al menu principal \n''' ))
  return a
```

```python
## bloque de operaciones basicas que contiene las funciones que se llaman según las opciones elegidas

def get_suma():
  print("\n Suma de dos numeros reales (x = y + z)\n")
  a = float(input("ingrese y"))
  b = float(input("ingrese z"))
  print(a + b)

def get_resta():
  print("\n resta de dos numeros reales (x = y - z)\n")
  a = float(input("ingrese y"))
  b = float(input("ingrese z"))
  print(a - b)

def get_multiplicacion():
  print("\n multiplicacion de dos numeros reales (x = y * z)\n")
  a = float(input("ingrese y"))
  b = float(input("ingrese z"))
  print(a * b)

def get_division():
  print("\n division de dos numeros reales (x = y / z)\n")
  a = float(input("ingrese y"))
  b = float(input("ingrese z"))

  if b == 0:
    print("Usted ingresó z = 0 lo cual no está permitido")
  else:
    print(a / b)
```

```python
## bloque de operaciones extendidas que contiene las funciones que se llaman según las opciones elegidas

import math

def get_entera():
  print("\n regresa la parte entera de la division de dos numeros reales {x = int( y / z)}\n")
  a = float(input("ingrese y ="))
  b = float(input("ingrese z ="))

  if b == 0: print("Usted ingresó z = 0 lo cual no está permitido")
  else: print(a // b)

def get_residuo():
  print("\n regresa el residuo de la division de dos numeros reales x = residuo de ( y / z) \n")
  a = float(input("ingrese y ="))
  b = float(input("ingrese z ="))

  if b == 0: print("Usted ingresó z = 0 lo cual no está permitido")
  else: print(a % b)

def get_expo():
  print("\n regresa el resultado de y**z\n")
  a = float(input("ingrese y ="))
  b = float(input("ingrese z ="))
  print(a ** b)

def get_raiz():
  print("\n regresa la raiz indicada x de un número real y \n")
  a = float(input("ingrese x ="))
  b = float(input("ingrese y ="))

  print(b**(1/a))

def get_logn():
  print("\n regresa el logaritmo natural de un número real x \n")
  a = float(input("ingrese x ="))

  print(math.log(a))

def get_log10():
  print("\n regresa el logaritmo base 10 de un número real x \n")
  a = float(input("ingrese x ="))

  print(math.log10(a))

def get_inversa():
  print("\n regresa la inversa de un número real x \n")
  a = float(input("ingrese x ="))

  print(1/a)

def get_inversa():
  print("\n regresa la inversa de un número real x \n")
  a = float(input("ingrese x ="))

  print(1/a)

def get_factorial():
  print("\n regresa el factorial de un número entero x \n")
  a = int(input("ingrese x ="))

  fact = 1
  for i in range(1,a+1,1):
    fact *= i
  print(fact)

def get_abs():
  print("\n regresa el valor absoluto de un número real x \n")
  a = float(input("ingrese x ="))

  print(math.fabs(a))

```

```python
## bloque de operaciones trigonometricas que contiene las funciones que se llaman según las opciones elegidas

def get_seno():
  print("\n regresa el seno de un número real x \n")
  a = float(input("ingrese x ="))

  print(math.sin(a))

def get_tan():
  print("\n regresa la tangente de un número real x \n")
  a = float(input("ingrese x ="))

  print(math.tan(a))

def get_cos():
  print("\n regresa el coseno de un número real x \n")
  a = float(input("ingrese x ="))

  print(math.cos(a))
```

```python
## bloque de operaciones estadisticas que contiene las funciones que se llaman según las opciones elegidas

import numpy as np
from scipy import stats

def get_promedio():
  print("\n regresa el promedio de un conjunto X de números reales donde X = [x1, x2, x3, ..., xn] \n")
  a = input("ingrese los elementos del conjunto separados por coma")
  a = a.split(',')
  float_array = [float(i) for i in a]
  print(np.mean(float_array))

def get_mediana():
  print("\n regresa la mediana de un conjunto X de números reales donde X = [x1, x2, x3, ..., xn] \n")
  a = input("ingrese los elementos del conjunto separados por coma")
  a = a.split(',')
  float_array = [float(i) for i in a]

  print(np.median(np.sort(float_array)))

def get_moda():
  print("\n regresa la moda de un conjunto X de números reales donde X = [x1, x2, x3, ..., xn] \n")
  a = input("ingrese los elementos del conjunto separados por coma")
  a = a.split(',')
  float_array = [float(i) for i in a]

  print(stats.mode(float_array))
```

```python
#menu principal
show_menu = True

while show_menu == True:

  menu_operations = menu_options() #aquí llama la funcion que captura por consola la opcion de tipo de operacion deseada

      # si selecciona basica
  if menu_operations == 1:
    menu_basica = True

    while menu_basica == True:

      sub_basica = basic_options() # se ejecuta el ciclo del submenu basica que captura por consola la operacion deseada

      if sub_basica == 1: get_suma()
      elif sub_basica == 2: get_resta()
      elif sub_basica == 3: get_multiplicacion()
      elif sub_basica == 4: get_division()
      elif sub_basica == 5: break
      else: print("opcion inválida")

    # si selecciona extendida
  elif menu_operations == 2:
    menu_extendida = True

    while menu_extendida == True:

      sub_extendida = extended_options()  # se ejecuta el ciclo del submenu extendida que captura por consola la operacion

      if sub_extendida == 1: get_entera()
      elif sub_extendida == 2: get_residuo()
      elif sub_extendida == 3: get_expo()
      elif sub_extendida == 4: get_raiz()
      elif sub_extendida == 5: get_logn()
      elif sub_extendida == 6: get_log10()
      elif sub_extendida == 7: get_inversa()
      elif sub_extendida == 8: get_factorial()
      elif sub_extendida == 9: get_abs()
      elif sub_extendida == 10: break
      else: print("opcion inválida")

# si selecciona trigonometrica
  elif menu_operations == 3:
   menu_trigo = True

   while menu_trigo == True:

      sub_trigo = trigo_options()  # se ejecuta el ciclo del submenu trigonometrica que captura por consola la operacion deseada

      if sub_trigo == 1: get_seno()
      elif sub_trigo == 2: get_tan()
      elif sub_trigo == 3: get_cos()
      elif sub_trigo == 4: break
      else: print("opcion inválida")

# si selecciona  estadistica

  elif menu_operations == 4:
   menu_est = True

   while menu_est == True:

      sub_est = estadistica_options()  # se ejecuta el ciclo del submenu estadistica que captura por consola la operacion deseada

      if sub_est == 1: get_promedio()
      elif sub_est == 2: get_mediana()
      elif sub_est == 3: get_moda()
      elif sub_est == 4: break
      else: print("opcion inválida")

  else:  #si oprime otro numero que no está en la lista de opciones entonces finaliza el ciclo principal del menu y termina el programa
    print(''' ¡Hasta pronto! ''')
    break
```

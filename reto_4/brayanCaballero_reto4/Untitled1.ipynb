```python
# Buscamos una bibliotecas que nos sirva, en este caso:

import math

# En estas lineas, definiremos la clase. En este caso, una calculadora para así crear caracteristicas de la misma. Por otro lado, encapsula todas las funciones y datos relacionados con la calculadora #
class Calculator:
    def __init__(self):
        self.result = None

# Función para imprimir el menú principal


    def print_main_menu(self):
        print("Calculadora científica")
        print("1. Operaciones aritméticas básicas")
        print("2. Operaciones aritméticas extendidas")
        print("3. Operaciones trigonométricas")
        print("4. Operaciones estadísticas básicas")
        print("5. Salir")

# Función para imprimir y realizar operaciones aritmeticas

    def arithmetic_operations(self):
        try:
            num1 = float(input("Ingrese el primer número: "))
            num2 = float(input("Ingrese el segundo número: "))
        except ValueError:
            print("Entrada no válida. Asegúrate de ingresar números.")
            return None

        print("Seleccione la operación aritmética que desea realizar:")
        print("1. Suma")
        print("2. Resta")
        print("3. Multiplicación")
        print("4. División")
        operation = input("Ingrese su selección: ")

    # Validar la selección del menú

        if operation not in ("1", "2", "3", "4"):
            print("Opción no válida.")
            return None
    # Función para realizar las operaciones

        if operation == "1":
            return num1 + num2
        elif operation == "2":
            return num1 - num2
        elif operation == "3":
            return num1 * num2
        elif operation == "4":
            if num2 == 0:
                print("No se puede dividir por cero.")
                return None
            return num1 / num2

# Función para imprimir y realizar operaciones aritmeticas extendidas

    def extended_arithmetic_operations(self):
        print("Seleccione la operación aritmética extendida que desea realizar:")
        print("1. División entera")
        print("2. Residuo")
        print("3. Exponenciación a la n")
        print("4. Raíz n")
        print("5. Logaritmo en base 10")
        print("6. Logaritmo natural")
        print("7. Valor absoluto")
        print("8. 1/x")
        print("9. Factorial")
        operation = input("Ingrese su selección: ")

  # Validar la selección del menú

        if operation not in ("1", "2", "3", "4", "5", "6", "7", "8", "9"):
            print("Opción no válida.")
            return None

        try:
            num = float(input("Ingrese el número: "))
        except ValueError:
            print("Entrada no válida. Asegúrate de ingresar un número.")
            return None

  # Función para realizar las operaciones

        if operation == "1":
            return math.floor(num)
        elif operation == "2":
            return num % 1
        elif operation == "3":
            return math.pow(num, float(input("Ingrese el exponente: ")))
        elif operation == "4":
            return math.sqrt(num)
        elif operation == "5":
            return math.log10(num)
        elif operation == "6":
            return math.log(num)
        elif operation == "7":
            return abs(num)
        elif operation == "8":
            if num == 0:
                print("No se puede dividir por cero.")
                return None
            return 1 / num
        elif operation == "9":
            if num < 0 or not num.is_integer():
                print("El número debe ser un entero no negativo.")
                return None
            return math.factorial(int(num))

# Define un método para ejecutar la calculadora. Presenta el menú principal, recopila la elección del usuario y ejecuta la operación correspondiente.

    def run(self):
        while True:
            self.print_main_menu()
            choice = input("Seleccione una opción (1-5): ")

            if choice == "5":
                print("¡Hasta luego!")
                break
            elif choice == "1":
                self.result = self.arithmetic_operations()
            elif choice == "2":
                self.result = self.extended_arithmetic_operations()

            if self.result is not None:
                print("El resultado es:", self.result)


# Instanciamos la calculadora y la ejecutamos
calculator = Calculator()
calculator.run()
```

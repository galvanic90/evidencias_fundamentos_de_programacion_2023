Punto 2.  Desarrolle una calculadora científica que permita realizar las siguientes operaciones:

•	Operaciones aritméticas básicas: las operaciones aritméticas básicas son suma, resta, multiplicación, división.
•	Operaciones aritméticas extendidas: son la división entera, el residuo, la exponenciación a la n de número, raíz n (n puede ser 2, 3, 4, 5) de un número, valor absoluto, 1/número, factorial.
•	Operaciones trigonométricas: al menos dos, por ejemplo, seno(x), tangente(x).
•	Operaciones estadísticas básicas: promedio, media, mediana, moda.  


```python
import math  # Importamos la biblioteca math para las operaciones trigonométricas y raíces

# Función para mostrar el menú principal
def mostrar_menu_principal():
    print("CALCULADORA CIENTÍFICA")
    print("1. Operaciones Aritméticas")
    print("2. Operaciones Aritméticas Extendidas")
    print("3. Operaciones Trigonométricas")
    print("4. Operaciones Estadísticas")
    print("5. Salir")

# Función para realizar operaciones aritméticas básicas
def operaciones_aritmeticas():
    print("OPERACIONES ARITMÉTICAS BÁSICAS")
    num1 = float(input("Ingrese el primer número: "))
    num2 = float(input("Ingrese el segundo número: "))

    print("Resultado:")
    print("Suma:", num1 + num2)
    print("Resta:", num1 - num2)
    print("Multiplicación:", num1 * num2)

    if num2 != 0:
        print("División:", num1 / num2)
    else:
        print("División: No se puede dividir por cero.")

# Función para realizar operaciones aritméticas extendidas
def operaciones_aritmeticas_extendidas():
    print("OPERACIONES ARITMÉTICAS EXTENDIDAS")
    num = float(input("Ingrese un número: "))

    print("Resultado:")
    print("División Entera por 2:", num // 2)
    print("Residuo por 2:", num % 2)
    exponente = int(input("Ingrese el exponente (n): "))
    print("Exponenciación a la n:", num ** exponente)

    n_raiz = int(input("Ingrese el valor de n para la raíz (2, 3, 4, 5): "))
    if n_raiz >= 2:
        print(f"Raíz {n_raiz}-ésima:", num ** (1 / n_raiz))
    else:
        print("Valor de n no válido para la raíz.")

    print("Valor Absoluto:", abs(num))
    if num != 0:
        print("1 / Número:", 1 / num)
    else:
        print("1 / Número: No se puede dividir por cero.")

    factorial = math.factorial(int(num))
    print("Factorial:", factorial)

# Función para realizar operaciones trigonométricas
def operaciones_trigonometricas():
    print("OPERACIONES TRIGONOMÉTRICAS")
    angulo = float(input("Ingrese el valor del ángulo en grados: "))

    angulo_rad = math.radians(angulo)  # Convertir a radianes para funciones trigonométricas

    print("Resultado:")
    print("Seno:", math.sin(angulo_rad))
    print("Tangente:", math.tan(angulo_rad))

# Función para realizar operaciones estadísticas básicas
def operaciones_estadisticas():
    print("OPERACIONES ESTADÍSTICAS")
    numeros = input("Ingrese una lista de números separados por espacios: ").split()
    numeros = [float(num) for num in numeros]

    print("Resultado:")
    print("Promedio:", sum(numeros) / len(numeros))
    print("Media:", sorted(numeros)[len(numeros) // 2])
    from collections import Counter
    moda = [item for item, count in Counter(numeros).items() if count == max(Counter(numeros).values())]
    print("Moda:", moda)

# Función principal
def main():
    while True:
        mostrar_menu_principal()
        opcion = input("Elija una opción (1/2/3/4/5): ")

        if opcion == '1':
            operaciones_aritmeticas()
        elif opcion == '2':
            operaciones_aritmeticas_extendidas()
        elif opcion == '3':
            operaciones_trigonometricas()
        elif opcion == '4':
            operaciones_estadisticas()
        elif opcion == '5':
            print("¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, elija una opción válida.")

# Llamar a la función principal
if __name__ == "__main__":
    main()
```

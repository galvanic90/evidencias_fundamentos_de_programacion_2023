```python
#Libreria que me permite realizar calculos matematicos
import math

# Función para Suma
def suma(a, b):
    return a + b
# Función para resta
def resta(a, b):
    return a - b

# Función para multiplicacion
def multiplicacion(a, b):
    return a * b

# Función para dividir
def division(a, b):
    if b == 0:
        return "No se puede dividir por cero"
    else:
        return a / b

#PARTE B
# Función para la división entera
def division_entera(a, b):
    return a // b
# Función para obtener el residuo
def residuo(a, b):
    return a % b
# Función para la exponenciación a la n de un número
def exponente(a, n):
    return a ** n
# Función para la raíz n de un número
def raiz_n(a, n):
    return a ** (1/n)

# Función para el logaritmo en base 10
def logaritmo_base_10(a):
    return math.log10(a)

# Función para el logaritmo natural (base e)
def logaritmo_natural(a):
    return math.log(a)


# Función para obtener el valor absoluto
def valor_absoluto(a):
    return abs(a)

# Función para calcular 1/número
def inverso(a):
    if a == 0:
        return "No se puede calcular el inverso de cero"
    else:
        return 1 / a

# Función para calcular el factorial
def factorial(n):
    if n < 0:
        return "No se puede calcular el factorial de un número negativo"
    elif n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Función para calcular el seno de un ángulo en grados
def seno(angulo_grados):
    angulo_radianes = math.radians(angulo_grados)
    return math.sin(angulo_radianes)

# Función para calcular la tangente de un ángulo en grados
def tangente(angulo_grados):
    angulo_radianes = math.radians(angulo_grados)
    return math.tan(angulo_radianes)
# Función principal de la calculadora
def calculadora():
    print("Calculadora de Operaciones Aritméticas")
    while True:
        print("Operaciones Aritmeticas:")
        print("1. Suma")
        print("2. Resta")
        print("3. Multiplicación")
        print("4. División")
        print("5. Division Entera")
        print("6. Residuo")
        print("7. Exponente")
        print("8. raiz_n")
        print("9. logaritmo_base_10")
        print("10. logaritmo_natural")
        print("11. valor_absoluto")
        print("12. inverso ")
        print("13. factorial")
        print("14. seno")
        print("15. tangente")
        print("16. Salir")

        opcion = input("Seleccione una operación (): ")

        if opcion == '16':
            break
         #En este bloque de codigo solo requerimos utlizar solo 1 numero para la funciones descritas
        if opcion in ('9','10','11','12','13','14','15'):
          num4 = float(input("Ingrese el primer número: "))
          if opcion == '9':
           print("Resultado:", logaritmo_base_10(num4))
          elif opcion == '10':
            print("Resultado:", logaritmo_natural(num4))
          elif opcion == '11':
            print("Resultado:", valor_absoluto(num4))
          elif opcion == '12':
            print("Resultado:", inverso(num4))
          elif opcion == '13':
            print("Resultado:", factorial(num4))
          elif opcion == '14':
            print("Resultado:", seno(num4))
          elif opcion == '15':
            print("Resultado:", tangente(num4))
          #En este bloque de codigo solo requerimos utlizar solo 2 numero para la funciones descritas
        if opcion in ('1', '2', '3', '4','5','6','7','8'):
            num1 = float(input("Ingrese el primer número: "))
            num2 = float(input("Ingrese el segundo número: "))


            if opcion == '1':
                print("Resultado:", suma(num1, num2))
            elif opcion == '2':
                print("Resultado:", resta(num1, num2))
            elif opcion == '3':
                print("Resultado:", multiplicacion(num1, num2))
            elif opcion == '4':
                print("Resultado:", division(num1, num2))
            elif opcion == '5':
                print("Resultado:", division_entera(num1, num2))
            elif opcion == '6':
                print("Resultado:", residuo(num1, num2))
            elif opcion == '7':
                print("Resultado:", exponente(num1, num2))
            elif opcion == '8':
                print("Resultado:", raiz_n(num1, num2))

            else:
             print("Opción no válida. Intente de nuevo.")


if __name__ == "__main__":
    calculadora()
```

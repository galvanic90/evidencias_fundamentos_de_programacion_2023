```python
import math
from statistics import mean, median, mode

# Definición de funciones para las operaciones aritméticas básicas
def suma(a, b):     # Suma
    return a + b

def resta(a, b):    # Resta
    return a - b

def multiplicacion(a, b):   #Multiplicacion
    return a * b

def division(a, b):     #Division
    if b != 0:      #Aquí tengo en cuenta el hecho de que la división por cero daría error
        return a / b
    else:
        return "Error: división por cero"

# Definición de funciones para las operaciones aritméticas extendidas

def division_entera(a, b):  #Division entera
    if b != 0:      #Aquí tengo en cuenta el hecho de que la división por cero daría error
        return a // b
    else:
        return "Error: división por cero"

def residuo(a, b):  #Residuo de la division
    if b != 0:      #Aquí tengo en cuenta el hecho de que la división por cero daría error
        return a % b
    else:
        return "Error: división por cero"

def potencia(a, n):     #Potenciacion
    return math.pow(a, n)

def raiz(a, n):     #Radicalización
    if a >= 0:      #Aquí tengo en cuenta que no existen raíces reales para los números negativos, esto daría números complejos
        return math.pow(a, 1/n)
    else:
        return "Error: no se puede calcular la raíz de un número negativo"

def logaritmo_base_10(a):   #Logaritmo en base 10
    return math.log10(a)

def logaritmo(a, base):     #Logaritmo en cualquier base
    return math.log(a, base)

def valor_absoluto(a):      #Valor absoluto
    return abs(a)

def inverso(a):     #Inverso del número
    if a != 0:      #Aquí tengo en cuenta el hecho de que la división por cero daría error
        return 1 / a
    else:
        return "Error: división por cero"

def factorial(a):       #Factorial del número
    if a >= 0:      #El factorial de un número está definido para los números enteros naturales
        return math.factorial(a)
    else:
        return "Error: no se puede calcular el factorial de un número negativo"

# Definición de funciones para las operaciones trigonométricas

def seno(x):    #Función trigonométrica seno
    return math.sin(x)

def tangente(x):    #Función trigonométrica tangente
    return math.tan(x)

# Definición de función para las operaciones estadísticas básicas

def calculadora_estadistica(datos):   #Función para calcular los valores estadísticos
    promedio = mean(datos)      #Cálculo del promedio
    mediana = median(datos)     #Cálculo de la mediana
    moda = mode(datos)          #Cálculo de la moda
    return promedio, mediana, moda

# Ciclo infinito para el menú principal de la calculadora
while True:
    print("Calculadora Científica")
    print("1. Operaciones aritméticas básicas")
    print("2. Operaciones aritméticas extendidas")
    print("3. Operaciones trigonométricas")
    print("4. Operaciones estadísticas básicas")
    print("5. Salir")

    opcion = input("Seleccione una opción (1/2/3/4/5): ")

    if opcion == '1':
        a = float(input("Ingrese el primer número: "))
        b = float(input("Ingrese el segundo número: "))
        print("Suma:", suma(a, b))
        print("Resta:", resta(a, b))
        print("Multiplicación:", multiplicacion(a, b))
        print("División:", division(a, b))
        print()
    elif opcion == '2':
        a = float(input("Ingrese un número: "))
        n = int(input("Ingrese el valor de n: "))
        print("División entera:", division_entera(a, n))
        print("Residuo:", residuo(a, n))
        print("Potencia:", potencia(a, n))
        print("Raíz", n, "de", a, ":", raiz(a, n))
        print("Logaritmo en base 10:", logaritmo_base_10(a))
        base = float(input("Ingrese la base del logaritmo: "))
        print("Logaritmo en base", base, ":", logaritmo(a, base))
        print("Valor absoluto:", valor_absoluto(a))
        print("Inverso:", inverso(a))
        print("Factorial:", factorial(int(a)))
        print()
    elif opcion == '3':
        x = float(input("Ingrese un ángulo en radianes: "))
        print("Seno:", seno(x))
        print("Tangente:", tangente(x))
        print()
    elif opcion == '4':
        datos = input("Ingrese una lista de números separados por comas (ejemplo: 1,2,3): ")
        datos = [float(x.strip()) for x in datos.split(',')]
        promedio, mediana, moda = calculadora_estadistica(datos)
        print("Promedio:", promedio)
        print("Mediana:", mediana)
        print("Moda:", moda)
        print()
    elif opcion == '5':
        break
    else:
        print("Opción no válida. Por favor, seleccione una opción válida.")
```
